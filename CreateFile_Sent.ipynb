```python
import pandas as pd
import numpy as np

dc_listings = pd.read_csv('listings.csv')
dc_reviews = pd.read_csv('reviews.csv')

merged = pd.merge(dc_listings, dc_reviews, on='id', how='inner')
```

```python
merged.to_csv("merged_master.csv")
```

```python
merged['id'].nunique()
```

```python
merged.describe()
```

```python
from langdetect import detect
def detectlang(comment):
  try:
    return detect(comment)
  except:
    return 0

merged['detect'] = merged['comments'].apply(lambda comment: detectlang(comment = comment))
merged.head()
```

```python
merged.to_csv("merged_with_lang.csv")
```

```python
final = merged[merged["detect"] == "en"]
final.to_csv("merged_with_en_comments.csv")
```

```python
final['comments'] = final['comments'].str.replace(r'[^\d.a-zA-Z]+', " ")
final['comments'] = final['comments'].str.casefold()
final.head()
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

all = []
for review in final.comments:
    sid = SentimentIntensityAnalyzer()
    scores = sid.polarity_scores(review)
    all.append(scores['compound'])

final['sentiment'] = all
```

```python
final.to_csv("en_sent_merged.csv")
```
