```python
import pandas as pd
import numpy as np

dc_listings = pd.read_csv('main2.csv')
```

```python
dc_listings.head()
```

```python
final = dc_listings[dc_listings["detect"] == "en"]
```

```python
final.detect.unique()
```

```python
final.shape
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

all = []
for review in final.comments:
    sid = SentimentIntensityAnalyzer()
    scores = sid.polarity_scores(review)
    all.append(scores['compound'])

final['sentiment'] = all
```

```python
final.to_csv("newfinal.csv")
```
